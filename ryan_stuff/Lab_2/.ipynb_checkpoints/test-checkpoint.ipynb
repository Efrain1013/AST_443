```python
from astropy.io import fits
import matplotlib.pyplot as plt

image = fits.open('../../Lab_2/images/lab2_science_red_20s.00000060.FIT')

data = image[0].data

print(data.shape)
plt.imshow(data,cmap='gray',vmax=1500)
plt.colorbar()
```

```python
flat_data = data[0].flatten()

plt.hist(flat_data,log=True,bins=200)
plt.show()
```

```python
mask = flat_data < 1400

filtered_data = flat_data[mask]

plt.hist(filtered_data,log=True,bins=200)
plt.show()
```
