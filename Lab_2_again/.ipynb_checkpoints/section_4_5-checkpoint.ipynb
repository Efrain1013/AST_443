```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

JDs = []
flux = []
flux_err = []

with open("NEW_SECTION_4_3/kepler1.dat") as file:
    lines = file.readlines()
    column0 = []
    column1 = []
    column2 = []
    for x in lines:
        column0.append(float(x.split(' ')[0]))
        column1.append(float(x.split(' ')[1]))
        column2.append(float(x.split(' ')[2]))
    JDs.append(column0)
    flux.append(column1)
    flux_err.append(column2)

for i in range(1, 11):
    filename = "NEW_SECTION_4_3/ref" + str(i) + ".dat"
    with open(filename, 'r') as file:
        lines = file.readlines()
        column0 = []
        column1 = []
        column2 = []
        for x in lines:
            column0.append(float(x.split(' ')[0]))
            column1.append(float(x.split(' ')[1]))
            column2.append(float(x.split(' ')[2]))
        JDs.append(column0)
        flux.append(column1)
        flux_err.append(column2)
        
# The list corresponding to our last reference star does not have the last time on the measuremnet
# Rather than try to come up with some data on that last part, I cut out the last point on the other lists
for i in range(0, 10):
    JDs[i].pop(-1)
    flux[i].pop(-1)
    flux_err[i].pop(-1)

```

```python
index = 0
while index < 11:
    j = 0
    for j in range(0, len(flux[index][:])):
        if j >= len(flux[index][:]):
            break
        if flux[index][j] == 0.001 or flux[index][j] < 0.:
            JDs[index].pop(j)
            flux[index].pop(j)
            flux_err[index].pop(j)
            j -= 1
    if 0.001 in flux[index]:
        index -= 1
    index += 1

#2460347.489977
```

```python
for i in range(0, 11):
#     plt.figure(figsize=(20, 10))
    plt.xlabel('Time (JD)')
    plt.plot(JDs[i], flux[i])
    plt.errorbar(JDs[i], flux[i], linestyle='', c='r', yerr=flux_err[i])
    plt.ylabel('Flux')
    if i == 0:
        plt.title("Keppler 1")
    else:
        plt.title(rf'REF {i}')
    plt.show()
```

```python
# Here we want to line up our data so that when we calculate the weighted means we're not
# mixing our points together

ind_targ = ind_1 = ind_2 = ind_2 = ind_3 = ind_4 = ind_5 = ind_6 = ind_7 = ind_8 = ind_9 = ind_10 = 0
ind = [ind_targ, ind_1, ind_2, ind_3, ind_4, ind_5, ind_6, ind_7, ind_8, ind_9, ind_10]

weighted_means = []
weighted_errs = []

while np.max(ind) < len(JDs[0]):
    temp_flux = []
    temp_flux_err = []
    temp_flux = np.array(temp_flux)
    temp_flux_err = np.array(temp_flux_err)
    date = JDs[0][ind[0]]

    for i in range(1, 11):
        if ind[i] >= len(JDs[i]):
            break
        if JDs[i][ind[i]] == date:
            temp_flux = np.append(temp_flux, flux[i][ind[i]])
            temp_flux_err = np.append(temp_flux_err, flux_err[i][ind[i]])
            ind[i] += 1

    ind[0] += 1
    
    if len(temp_flux_err) != 0:
        w_mean = sum(temp_flux / (temp_flux_err**2)) / sum(1. / temp_flux_err**2)
        w_err = np.sqrt(sum(1. / temp_flux_err**2))
    else:
        w_mean = 1.0
        w_err = 1.0
    weighted_means.append(w_mean)
    weighted_errs.append(w_err)
#     print(f"Weighted mean is {w_mean}")
```

```python
for i in range(0, 11):
    print(JDs[i][-1])

r_s = []
r_s_err = []
for i in range(0, len(flux[0])):
    r_s.append(flux[0][i] / weighted_means[i])
    r_s_err.append(np.abs(flux[0][i] / weighted_means[i]) * 
                   np.sqrt((flux_err[0][i]/flux[0][i])**2 + (weighted_errs[i]/weighted_means[i])*2))
```

```python
pre_transit = 15

pre_flux = r_s[:pre_transit]
median_flux = np.median(pre_flux)
flux_norm = r_s / median_flux
flux_norm_err = r_s_err / median_flux
```

```python
plt.plot(JDs[0], flux_norm,'o')
plt.errorbar(JDs[0], flux_norm, linestyle='', c='r', yerr=flux_norm_err)
plt.ylim(0.8, 1.2)
plt.show()

```

```python
from scipy import optimize
import batman

time = np.asarray(JDs[0])-JDs[0][0]

def batman_model(time, t0, per, rp, a, inc, ecc, w, u1, u2):
#     t0, per, rp, a, inc, ecc, w, u1, u2 = params

    # Transit parameters
    transit_params = batman.TransitParams()
    transit_params.t0 = t0
    transit_params.per = per
    transit_params.rp = rp
    transit_params.a = a
    transit_params.inc = inc
    transit_params.ecc = ecc
    transit_params.w = w
    transit_params.u = [u1, u2]
    transit_params.limb_dark = "quadratic"
    
    # Transit model
    transit_model = batman.TransitModel(transit_params, time)
    model_flux = transit_model.light_curve(transit_params)

    return model_flux

# Initial guess for parameters
initial_guess = [0.05, 0.5, 0.1, 15.0, 87.0, 0.0, 90.0, 0.1, 0.3]

for name in ["lm", "trf", "dogbox"]:
  
    optimized_params, covariance = optimize.curve_fit(batman_model, time, flux_norm, p0=initial_guess, sigma = flux_norm_err, method=name)
    
    errors = np.sqrt(np.diag(covariance))
    
    transit_params = batman.TransitParams()
    transit_params.t0 = optimized_params[0]
    t0_err = errors[0]
    transit_params.per = optimized_params[1]
    per_err = errors[1]
    transit_params.rp = optimized_params[2]
    rp_err = errors[2]
    transit_params.a = optimized_params[3]
    a_err = errors[3]
    transit_params.inc = optimized_params[4]
    inc_err = errors[4]
    transit_params.ecc = optimized_params[5]
    ecc_err = errors[5]
    transit_params.w = optimized_params[6]
    w_err = errors[6]
    transit_params.u = [optimized_params[7], optimized_params[8]]
    u_err = [errors[7], errors[8]]
    transit_params.limb_dark = "quadratic"

    transit_model = batman.TransitModel(transit_params, time)
    model_flux = transit_model.light_curve(transit_params)

    # Print optimized parameters
    print(name, "Method\n")
    print(f"Mid Transit Time:       {optimized_params[0]:.5f} +/- {t0_err:.5f} days")
    print(f"Period:                 {optimized_params[1]:.5f} +/- {per_err:.5f} days")
    print(f"Planet Radius (model):  {optimized_params[2]* 9.73116:.5f} +/- {rp_err* 9.73116:.5f} Jupiter Radii")
    print(f"Transit Depth:          {1.-np.min(model_flux):.5f}")
    print(f"Planet Radius:          {np.sqrt(1-np.min(model_flux)) * 9.73116:.5f} Jupiter Radii")
    
    plt.figure(figsize=(10, 6))
    for i in range(0, len(model_flux)):
        if model_flux[i] != 1.0:
            plt.axvline(time[i], c='r', linestyle='--', alpha = 0.5)
            transit_start = time[i]
            break
    for i in range(len(model_flux)-1, 0, -1):
        if model_flux[i] != 1.0:
            plt.axvline(time[i], c='r', linestyle='--', alpha = 0.5)
            transit_stop = time[i]
            break
            
    print(f"Transit Start:          {transit_start:.5f} days")
    print(f"Transit Stop:           {transit_stop:.5f} days")
    print(f"Transit Duration:       {transit_stop-transit_start:.5f} days")
    print(f"                        {(transit_stop-transit_start) * 24:.5f} hours")
    
    plt.axvline(transit_params.t0, c='r', linestyle='--', alpha = 0.5)
    plt.scatter(time, flux_norm, c='orange', label='Data')
    plt.plot(time, model_flux, c='blue', label='Model')
    plt.xlabel('Time (days)')
    plt.ylabel('Flux')
    plt.title('Transit Light Curve Model')
    
    plt.legend()
    plt.show()
    
    print("=" * 115)
```

```python
def chi_squared(params):
    t0, per, rp, a, inc, ecc, w, u1, u2 = params
    
    # Transit parameters
    transit_params = batman.TransitParams()
    transit_params.t0 = t0
    transit_params.per = per
    transit_params.rp = rp
    transit_params.a = a
    transit_params.inc = inc
    transit_params.ecc = ecc
    transit_params.w = w
    transit_params.u = [u1, u2]
    transit_params.limb_dark = "quadratic"
    
    # Transit model
    transit_model = batman.TransitModel(transit_params, time)
    model_flux = transit_model.light_curve(transit_params)
    
    # Calculate sum of squared differences
    return np.sum((flux_norm - model_flux)**2 / model_flux)

# Initial guess for parameters
initial_guess = [0.05, 0.5, 0.1, 15.0, 87.0, 0.0, 90.0, 0.1, 0.3]

for name in ["Nelder-Mead", "Powell", "CG", "BFGS", "L-BFGS-B"]:
    # Minimize squared difference to converge to parameters
    result = optimize.minimize(chi_squared, initial_guess, method=name)

    # Extract the parameters
    optimized_params = result.x

    # Plot observed data and model
    transit_params = batman.TransitParams()
    transit_params.t0 = optimized_params[0]
    transit_params.per = optimized_params[1]
    transit_params.rp = optimized_params[2]
    transit_params.a = optimized_params[3]
    transit_params.inc = optimized_params[4]
    transit_params.ecc = optimized_params[5]
    transit_params.w = optimized_params[6]
    transit_params.u = [optimized_params[7], optimized_params[8]]
    transit_params.limb_dark = "quadratic"

    transit_model = batman.TransitModel(transit_params, time)
    model_flux = transit_model.light_curve(transit_params)

    # Print optimized parameters
    print(name, "Method\n")
    print(f"Mid Transit Time:       {optimized_params[0]:.5f} days")
    print(f"Period:                 {optimized_params[1]:.5f} days")
    print(f"Planet Radius (model):  {optimized_params[2]*9.73116:.5f} Solar Radii")
    print(f"Transit Depth:          {1.-np.min(model_flux):.5f}")
    print(f"Planet Radius:          {np.sqrt(1-np.min(model_flux)) * 9.73116 * 1.003:.5f} Solar Radii")
    
    plt.figure(figsize=(10, 6))
    for i in range(0, len(model_flux)):
        if model_flux[i] != 1.0:
            plt.axvline(time[i], c='r', linestyle='--', alpha = 0.5)
            transit_start = time[i]
            break
    for i in range(len(model_flux)-1, 0, -1):
        if model_flux[i] != 1.0:
            plt.axvline(time[i], c='r', linestyle='--', alpha = 0.5)
            transit_stop = time[i]
            break
            
    print(f"Transit Start:          {transit_start:.5f} days")
    print(f"Transit Stop:           {transit_stop:.5f} days")
    print(f"Transit Duration:       {transit_stop-transit_start:.5f} days")
    print(f"                        {(transit_stop-transit_start) * 24:.5f} hours")
    
    plt.axvline(transit_params.t0, c='r', linestyle='--', alpha = 0.5)
    plt.scatter(time, flux_norm, c='orange', label='Data')
    plt.plot(time, model_flux, c='blue', label='Model')
    plt.xlabel('Time (days)')
    plt.ylabel('Flux')
    plt.title('Transit Light Curve Model')
    
    plt.legend()
    plt.show()
    
    print("=" * 115)
```
